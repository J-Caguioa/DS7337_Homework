<b>Extract noun phrase (NP) chunks from your reviews using the following procedure:

* In Python, use BeautifulSoup to grab the main review text from each link.  
* Next run each review text through a tokenizer, and then try to NP-chunk it with a shallow parser. 
* You probably will have too many unknown words, owing to proper names of characters, actors, and so on that are not in your working dictionary. Make sure the main names that are relevant to the movies in your collection of reviews are added to the working lexicon, and then run the NP chunker again.</b> <sub>[(back to top)](#toc)</sub>

```python
from bs4 import BeautifulSoup
import requests
import json
import nltk

def parse_html(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    return soup

# Unused due to need for try/except loop on pages without 'reviewBody'
def review_contents(soup):
    json_object = soup.select('script[type*="application/ld+json"]')
    reviewDict = json.loads(str(json_object[0].get_text()))
    movie = reviewDict['itemReviewed']['url']
    author = reviewDict['author']['name']
    rating = reviewDict['reviewRating']['ratingValue']
    review = reviewDict['reviewBody']
    return review
```

```python
review_texts = []

for link in review_permalinks:
    soup = parse_html(link)
    json_object = soup.select('script[type*="application/ld+json"]')
    reviewDict = json.loads(str(json_object[0].get_text()))
    try:
        review = reviewDict['reviewBody']
    except KeyError:
        continue
    review_texts.append(review)

review_texts
```

```python
reviews_tokenized = [nltk.word_tokenize(review) for review in review_texts]   
len(reviews_tokenized)
```
